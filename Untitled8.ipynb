```python
import datetime
from collections import defaultdict
import time

# Base URL of the reviews page (without the page number)
base_url = "https://www.rottentomatoes.com/m/1108372-fast_and_the_furious/reviews?type=user&page="

# Initialize session for persisting headers and cookies
session = requests.Session()

# Track the number of reviews scraped per year
reviews_per_year = defaultdict(int)

# Desired year range (2001‚Äì2012)
desired_year_range = range(2001, 2013)

# Track the number of reviews scraped in total
total_reviews_scraped = 0

# Loop through pages until we have 50 reviews for each year
page_num = 1
while True:
    url = base_url + str(page_num)
    print(f"Scraping page {page_num}...")

    # Send a GET request to fetch the page content
    response = session.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract reviews from the current page
        reviews = soup.find_all('div', class_='audience-review-row')

        if not reviews:
            print(f"No more reviews found on page {page_num}.")
            break  # Stop if no reviews are found

        for review in reviews:
            # Extract reviewer name
            reviewer_name = review.find('a', class_='audience-reviews__name')
            reviewer_name = reviewer_name.get_text(strip=True) if reviewer_name else 'Anonymous'

            # Extract review text
            review_text = review.find('p', class_='audience-reviews__review js-review-text')
            review_text = review_text.get_text(strip=True) if review_text else 'No review text available'

            # Extract review date
            review_date_str = review.find('span', class_='audience-reviews__duration')
            review_date_str = review_date_str.get_text(strip=True) if review_date_str else 'No date available'

            # Convert review date to datetime object
            try:
                review_date = datetime.datetime.strptime(review_date_str, '%b %d, %Y')  # Adjust the format as necessary
            except ValueError:
                review_date = None

            # Filter by year range (2001‚Äì2012)
            if review_date and review_date.year in desired_year_range:
                year = review_date.year
                if reviews_per_year[year] < 50:
                    # Extract rating score
                    rating = review.find('rating-stars-group')
                    rating_score = rating['score'] if rating else 'No rating available'

                    # Print the extracted data
                    print(f"Reviewer: {reviewer_name}")
                    print(f"Review Text: {review_text}")
                    print(f"Review Date: {review_date_str}")
                    print(f"Rating: {rating_score}")
                    print("="*50)

                    # Increment the count of reviews for this year
                    reviews_per_year[year] += 1
                    total_reviews_scraped += 1

                # Stop if we've scraped 50 reviews for all years
                if all(count >= 50 for count in reviews_per_year.values()):
                    print("Reached 50 reviews for each year in the desired range.")
                    break

        # Stop if we've scraped 50 reviews for all years
        if all(count >= 50 for count in reviews_per_year.values()):
            break

    else:
        print(f"Failed to retrieve page {page_num}. Status code: {response.status_code}")
        break

    # Go to the next page
    page_num += 1

    # Optional: Sleep between requests to avoid rate limiting
    time.sleep(2)

# Final message after scraping
print(f"Total reviews scraped: {total_reviews_scraped}")
```
